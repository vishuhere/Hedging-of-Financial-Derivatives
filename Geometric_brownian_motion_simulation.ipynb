# Geometric Brownian Motion (GBM) Simulation for Stock Prices

This notebook simulates stock prices using Geometric Brownian Motion (GBM). GBM is a foundational model in financial mathematics, widely used in pricing derivatives and modeling stock prices. This notebook provides a comprehensive tool for understanding and modeling the behavior of stock prices over time.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Parameters

- **S0**: Initial stock price
- **mu**: Drift term (average return)
- **sigma**: Volatility (standard deviation of returns)
- **T**: Time period (1 year in this example)
- **N**: Number of time steps (daily intervals, 252 days in a year)
- **dt**: Time step size, calculated as T / N
- **num_simulations**: Number of simulations to run

These parameters define the behavior of the stock price under the GBM model.


```python
# Parameters
S0 = 100     # Initial stock price
mu = 0.1     # Drift (average return)
sigma = 0.2  # Volatility (standard deviation of returns)
T = 1.0      # Time period (1 year)
N = 252      # Number of time steps (daily intervals)
dt = T / N   # Time step size
num_simulations = 1000  # Number of simulations
```

## Simulating GBM Paths

We will simulate multiple paths of the GBM to visualize the variation in possible stock price trajectories.


```python
# Time array
t = np.linspace(0, T, N)

# Simulating multiple GBM paths
S_paths = np.zeros((num_simulations, N))
for i in range(num_simulations):
    dW = np.random.normal(0, np.sqrt(dt), N)
    W = np.cumsum(dW)  # Cumulative sum to get the Wiener process
    S_paths[i, :] = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)
```

## Plotting Multiple GBM Paths

We plot several GBM paths to see the variation in stock price trajectories over time.


```python
# Plotting multiple GBM paths
plt.figure(figsize=(10, 6))
for i in range(num_simulations):
    plt.plot(t, S_paths[i, :], color='navy', alpha=0.1)
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
plt.title('Multiple Geometric Brownian Motion Simulations')
plt.grid(True)
plt.show()
```

## Mean and Median Path

We calculate and visualize the mean and median paths of the stock prices across all simulations for comparison.


```python
# Mean and Median Path
mean_path = np.mean(S_paths, axis=0)
median_path = np.median(S_paths, axis=0)

# Plotting Mean and Median Path
plt.figure(figsize=(10, 6))
plt.plot(t, mean_path, color='red', label='Mean Path')
plt.plot(t, median_path, color='green', label='Median Path')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
plt.title('Mean and Median Paths of GBM Simulations')
plt.legend()
plt.grid(True)
plt.show()
```

## Histogram of Final Stock Prices

We plot a histogram of the final stock prices after the time period to understand their distribution.


```python
# Histogram of final stock prices
final_prices = S_paths[:, -1]

plt.figure(figsize=(10, 6))
plt.hist(final_prices, bins=50, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Final Stock Price')
plt.ylabel('Frequency')
plt.title('Distribution of Final Stock Prices')
plt.grid(True)
plt.show()
```

## Boxplot of Final Stock Prices

We add a boxplot of the final stock prices to visualize the spread, median, and potential outliers.


```python
# Boxplot of final stock prices
plt.figure(figsize=(10, 6))
plt.boxplot(final_prices, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel('Final Stock Price')
plt.title('Boxplot of Final Stock Prices')
plt.grid(True)
plt.show()
```

## Relevance to Financial Derivatives

Geometric Brownian Motion (GBM) is widely used in financial mathematics, particularly in the Black-Scholes model for option pricing. The Black-Scholes model assumes that stock prices follow a GBM, which allows for the modeling of stock price movements and the calculation of option prices. Understanding GBM and its behavior is crucial for financial analysts and traders in making informed decisions regarding derivatives and risk management.
